```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
clf=LogisticRegression()
```

```python
t=pd.read_csv("train.csv")
t.head()
```

```python
del t["Name"]
t.head()
```

```python
del t["Ticket"]
t.head()
```

```python
del t["Cabin"]
t.head()
```

```python
del t["Fare"]
t.head()
```

```python
def change(str):
    if str=='male':
        return 1
    else:
        return 0
```

```python
t["Gender"]=t.Sex.apply(change)
t.head()
```

```python
t.drop("Sex",axis=1,inplace=True)
t.head()
```

```python
def change(port):
    if port=='S':
        return 1
    elif port=='Q':
        return 0
    return 2
```

```python
t["Port"]=t.Embarked.apply(change)
t.head()
```

```python
t.drop("Embarked",axis=1,inplace=True)
t.head()
```

```python
s_m=t[t.Survived==1].Age.mean()
s_m
```

```python
ns_m=t[t.Survived==0].Age.mean()
ns_m
```

```python
t["age"]=np.where(pd.isnull(t.Age) & t["Survived"]==1  ,s_m, t["Age"])
t
```

```python
t.age.fillna(ns_m,inplace=True)
t.head()
```

```python
t.drop("Age",axis=1,inplace=True)
t.head()
```

```python
t.rename(columns={"age":"Age"},inplace=True)
t.rename(columns={"Gender":"Sex"},inplace=True)
t.head()
```

```python
t["extra_col"] = t.iloc[:,3].copy()
t.head()
```

```python
t.drop("Survived",axis=1,inplace=True)
```

```python
t.rename(columns={"extra_col":"Survived"},inplace=True)
t.head()
```

```python
x_train=t.iloc[:,:-1]
y_train=t.iloc[:,-1]
```

```python
clf.fit(x_train,y_train)
```

```python
clf.score(x_train,y_train)
```
