```python
import numpy as np
```

```python
data1=np.genfromtxt("training_boston_x_y_train.csv",delimiter=",")
data2=np.genfromtxt("test_boston_x_test.csv",delimiter=",")
x_test=data2
x=data1[:,0:13]
y=data1[:,13]
print(x_test.shape)
print(x.shape)
```

```python
x=np.c_[x, np.ones(x.shape[0])]
x_test=np.c_[x_test, np.ones(x_test.shape[0])]
```

```python
def cost(s):
    total=0
    for i in range(x.shape[0]):
        a=x[i]
        b=y[i]
        total+=(1/x.shape[0])*((b-(s*a).sum())**2)
    return total
```

```python
def step_gradient(lr,s):
    M=len(x)
    ns=np.zeros(x.shape[1])
    for i in range(x.shape[0]):
        o=y[i]
        for j in range(x.shape[1]):
            c=x[i][j]
            ns[j]+=(-2/M)*(o-(s*x[i]).sum())*c
        s=s-lr*ns
    return s
```

```python
def gradient_descent():
    lr=0.0001
    n=1000000
    s=np.zeros(x.shape[1])
    for i in range(n):
        s=step_gradient(lr,s)
        print(i," cost: ",cost(s))
    return s
```

```python
slope=gradient_descent()
```

```python
print(slope)
```

```python
l=[]
l.append(slope)
l=np.array(l).reshape(-1,1)
print(l)
```

```python
y_pred=np.dot(x_test,slope)
```

```python
print(y_pred)
```
